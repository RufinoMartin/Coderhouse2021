```python
# Importacion Librerias

import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import pandas_profiling
from functools import reduce
import numpy as np
```

```python
# Leemos los datasets desde el Github de Oxford

data2020 = pd.read_csv('https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/OxCGRT_withnotes_2020.csv',low_memory=False,error_bad_lines = False)
data2021 = pd.read_csv('https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/OxCGRT_withnotes_2021.csv',low_memory=False,error_bad_lines = False)
data2022 = pd.read_csv('https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/OxCGRT_withnotes_2022.csv',low_memory=False,error_bad_lines = False)
```

```python
data2020.head(1)
```

```python
# Procedemos a dropear el grueso de las columnas irrelevantes de forma individual, por ser los datasets tan pesados.
# Dejamos las variables de interes, que son aquellos indices producidos por el equipo de investigacion,
# que nuclean 4 sectores de accion: 

#  'StringencyIndex': Indice de Rigor.
#  'GovernmentResponseIndex' : Indice de Respuesta Gubernamental.
#  'ContainmentHealthIndex' : Indice de Contencion. 
#  'EconomicSupportIndex' : Indice de Apoyo economico. 

#   En la documentacion se citan y desarrollan dichos indices, que utilizamos conjuntamente para nuestra variable:
#   "Restricciones por COVID19"
```

```python
# 2020

data2020.drop(data2020.columns.difference(['CountryName','Date','StringencyIndex','GovernmentResponseIndex',
                                    'ContainmentHealthIndex', 'EconomicSupportIndex',]), 1, inplace=True)
data2020.head(1)
data2020.to_csv('data2020.csv')
```

```python
# 2021

data2021.drop(data2021.columns.difference(['CountryName','Date','StringencyIndex','GovernmentResponseIndex',
                                    'ContainmentHealthIndex', 'EconomicSupportIndex',]), 1, inplace=True)
data2021.head(1)
data2021.to_csv('data2021.csv')
```

```python
# 2022

data2022.drop(data2022.columns.difference(['CountryName','Date','StringencyIndex','GovernmentResponseIndex',
                                    'ContainmentHealthIndex', 'EconomicSupportIndex',]), 1, inplace=True)
data2022.head(1)
data2022.to_csv('data2022.csv')
```

```python
# Borramos la columna 'unnam'e.

df_r2020.drop(columns={'Unnamed: 0'}, inplace=True)
df_r2021.drop(columns={'Unnamed: 0'}, inplace=True)
df_r2022.drop(columns={'Unnamed: 0'}, inplace=True)
```

```python
# Lo ideal seria, por cada dataframe de cada año (2020,2021 y 2022): 
#   -  obtener una sola fila por pais, con los cuatro indices promediados:
#         Aruba /2020 / primero.stringency / promedio.gresponse / promedio.chealth / promedio.esupport

#    De esa forma, conservamos informacion importante y resumimos el dataframe....lo que nos habilitará 
#       a hacer el merge luego...

df_r2020.head(12)
```

```python
# Procedemos a unir los dataframes

df_todos = [df_r2020, df_r2021, df_r2022]
```

```python
# Hacemos el merge utilizando la Fecha como clave principal.

df_merge = reduce(lambda  left,right: pd.merge(left,right,on=['CountryName'],
                                            how='outer'), df_todos)

df_merge.head.shape
```
